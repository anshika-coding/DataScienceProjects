#### <b>:- Next, weâ€™ll perform Exploratory Data Analysis (EDA) to dive into our data and understand it better. <b>Here, we visualize the average unemployment rates across different regions (states/UTs) to compare regional trends. Using a bar plot, we can easily see which regions have higher or lower unemployment. The colored bars and rotated labels make the chart more readable, helping us quickly identify patterns and regional differences in unemployment.<b>

```python
plt.figure(figsize=(15, 7))
color=["#FF9933", "#138808"]
sns.barplot(x="Region", y="Estimated Unemployment Rate (%)", data=unemp, ci=None,  palette=color)
plt.xticks(rotation=90)
plt.title("Unempolyement Rate by Region")
plt.xlabel("Reagion")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

<b>In this step, we use a box plot to explore the distribution and variation of labor participation rates across different regions. The plot highlights the spread, median, and potential outliers, giving us a clear view of how workforce participation differs regionally. This helps identify areas with unusually high or low participation rates.<b>

```python
plt.figure(figsize=(12, 6))
color=["#FF9933", "#138808"]
sns.boxplot(x="Region", y="Estimated Labour Participation Rate (%)", data=unemp, palette=color)
plt.title("Labor Participation Rate by Region")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

<b>Here, we calculate the average unemployment rate and labor participation rate for each region. By summarizing these key metrics, we can easily compare how different regions are performing and identify areas with higher unemployment or lower workforce participation.<b>

```python
avg_metrics = unemp.groupby('Region')[['Estimated Unemployment Rate (%)', 'Estimated Labour Participation Rate (%)']].mean()
```

```python
avg_metrics
```

<b>In this step, we create a bar chart to compare regions based on their average unemployment and labor participation rates. The plot makes it easy to see which regions are struggling with higher unemployment or lower workforce participation, providing a clear visual comparison across all regions.<b>

```python
avg_metrics.sort_values(by="Estimated Unemployment Rate (%)", ascending=False).plot(
    kind="bar", figsize=(12, 6),
    # Saffron & Green
    color=["#FF9933", "#138808"])
plt.title("Average Unemployment and Labor Participation Rate by Region", fontsize=14, fontweight="bold")
plt.ylabel("Percentage")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Convert the 'Date' column to datetime format  
# If any invalid date is found, it will be replaced with NaT (Not a Time) instead of raising an error  
unemp_rate['Date'] = pd.to_datetime(unemp_rate['Date'], errors='coerce',dayfirst=True)
```

```python
unemp_rate = unemp_rate.dropna(subset=['Date'])
```

<b>Filtering the dataset to only include records from the year 2020.
This helps in focusing on the unemployment trends during the Covid-19 period, allowing us to analyze how the pandemic impacted the labor market.<b>

```python
covid_data = unemp_rate[unemp_rate['Date'].dt.year == 2020]
```

```python
covid_data
```

### Q1: Visualizing the overall unemployment trend during COVID-19 in 2020  

**:- Using a line plot to show how unemployment rates changed over time**  
**:- 'hue=Region' adds separate lines for each region, making comparisons easier**  
**:- Grid is enabled for better readability of values**  

```python
plt.figure(figsize=(12, 6))
colors = ["#FF9933", "#FFD700", "#138808", "#000080", "#1CA9C9"]
sns.lineplot(data=covid_data, x='Date', y='Estimated Unemployment Rate', hue='Region', palette=colors)
plt.title("Trend of Unemployment Rate During COVID-19 (2020)")
plt.xlabel("Date")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.grid(True)
plt.show()
```

### Q2: Visualizing which regions faced the highest unemployment impact during COVID-19 (2020) 


**Using a boxplot to compare unemployment rate distribution across regions**

```python
plt.figure(figsize=(10, 6))
colors = ["#FF9933", "#138808","#000080", "#FF8000", "#006400"]  
sns.boxplot(data=covid_data, x='Region', y='Estimated Unemployment Rate', palette=colors)
plt.title("Unemployment Rate by Region During COVID-19 (2020)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Q3: Were there any seasonal trends in unemployment during the pandemic?


```python
covid_data['Month'] = covid_data['Date'].dt.month
monthly_avg = covid_data.groupby('Month')['Estimated Unemployment Rate'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Estimated Unemployment Rate', data=monthly_avg, marker='o', color="#E97451")
plt.title("Seasonal Trends in Unemployment Rate During COVID-19 (2020)")
plt.xlabel("Month")
plt.ylabel("Average Unemployment Rate (%)")
plt.grid(True)
plt.show()
```

### Q4: States most impacted by unemployment during COVID-19

```python
state_avg = covid_data.groupby('States')['Estimated Unemployment Rate'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=state_avg.values, y=state_avg.index,  palette='viridis')
plt.title("Top 10 States Most Impacted by Unemployment During COVID-19 (2020)")
plt.xlabel("Average Unemployment Rate (%)")
plt.ylabel("States")
plt.grid(True)
plt.show()
```
